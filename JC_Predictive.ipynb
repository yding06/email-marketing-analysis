```python
import pandas as pd
import numpy as np
```

```python
age22_campaign = pd.read_csv('age22_campaign.tsv', delimiter = '\t')
```

```python
age22_campaign['category_days_joined'] = age22_campaign['category_days_joined'].map({'(-0.362, 24.133]': 0, '(24.133, 48.267]': 1})
age22_campaign['category_hour'] = age22_campaign['category_hour'].map({'(-0.023, 5.75]': 0, '(5.75, 11.5]': 1})
age22_campaign['email_kind'] = age22_campaign['email_kind'].map({'job_listing': 0, 'cloud email': 1})
age22_campaign['weekday'] = age22_campaign['weekday'].map({'Monday': 0, 'Tuesday': 1, 'Wednesday':2,'Thursday':3,'Friday':4,'Saturday':5,'Sunday':6})
```

```python
age22_campaign = age22_campaign.drop(age22_campaign.columns[0],axis = 1)
age22_campaign = age22_campaign.drop('email_id',axis = 1)
```

```python
age22_campaign
```

```python
members = pd.read_csv('members_info.tsv', delimiter = '\t')
members = members.drop(members.columns[0], axis=1)
```

```python
df = age22_campaign.merge(members,how='left',on='member_id')
df = df.dropna(subset=['state'])
drop_elements = ['city', 'zip']
df = df.drop(drop_elements,axis =1)
df.loc[df['degree_level']=='\N','degree_level'] = np.nan
```

```python
categoryVarList = ['state','degree_level']
for var in categoryVarList:
    df[var] = df[var].astype('category')
df['opened_before'] =  df['opened_before'].astype(int)
df['age35'] =  df['age35'].astype(int)
```

```python
df['hs_or_ged_year'] = (df['hs_or_ged_year'] - df['hs_or_ged_year'].mean()) / (df['hs_or_ged_year'].max() - df['hs_or_ged_year'].min())
df
```

```python
df.to_csv('age22_campaign_pred.tsv', sep='\t')
```
