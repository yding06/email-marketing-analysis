```python
import pandas as pd
```

```python
email_sent_response = pd.read_csv('email_sent_response.tsv', delimiter = '\t')
members = pd.read_csv('members_info.tsv', delimiter = '\t')
```

```python
email_sent_response = email_sent_response.drop(email_sent_response.columns[0], axis=1)
members = members.drop(members.columns[0], axis=1)
```

```python
campaign_age22 = email_sent_response.loc[email_sent_response['age22']==1]
campaign_age22 = campaign_age22.drop_duplicates(subset=['member_id'],keep='last')
campaign_age22 = campaign_age22.drop(['opened_before'],axis=1)
campaign_age22
```

```python
campaign_opened = email_sent_response.loc[email_sent_response['opened_before']==1]
campaign_opened = campaign_opened.drop_duplicates(subset=['member_id'],keep='last')
opened = campaign_opened[['member_id','opened_before']].copy()
```

```python
temp0 = campaign_age22.merge(opened,how='left',on='member_id')
temp0 = temp0.drop_duplicates(subset=['member_id'],keep='last')
temp0 = temp0.drop(['age35'],axis=1)
```

```python
campaign_age35 = email_sent_response.loc[email_sent_response['age35']==1]
campaign_age35 = campaign_age35.drop_duplicates(subset=['member_id'],keep='last')
age35 = campaign_age35[['member_id','age35']].copy()
```

```python
temp1 = temp0.merge(age35,how='left',on='member_id')
drop_elements = ['email_type','open_time','click_time','click','unsub_time','unsub','age22','date']
temp1 = temp1.drop(drop_elements,axis=1)
temp1['opened_before']=temp1['opened_before'].fillna(0)
temp1['age35']=temp1['age35'].fillna(0)
temp1
```

```python
temp1.loc[ temp1['category_days_joined'] <= 24.133, 'category_days_joined'] = 0
temp1.loc[(temp1['category_days_joined'] > 24.133) & (temp1['category_days_joined'] <= 48.267), 'category_days_joined'] = 1
```

```python
temp1.to_csv('age22_campaign.tsv', sep='\t')
```
